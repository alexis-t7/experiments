```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()  # Create a new graph

# Add edges (nodes are automatically created)
G.add_edge(1, 2)
G.add_edge(1, 3)
G.add_edge(2, 4)
G.add_edge(3, 4)
```

```python
nx.draw(G, with_labels=True, node_color='lightblue', font_weight='bold')
plt.show()
```

```python
import numpy as np

# Adjacency Matrix
adj_matrix = nx.adjacency_matrix(G).todense()
print("Adjacency Matrix:")
print(adj_matrix)
```

```python
# Graph Laplacian L=D - A
laplacian_matrix = nx.laplacian_matrix(G).todense()
print("Graph Laplacian:")
print(laplacian_matrix)
```

```python
# Normalized Graph Laplacian L_normalized = D^(-1/2) * L * D^(-1/2)
norm_laplacian_matrix = nx.normalized_laplacian_matrix(G).todense()
print("Normalized Graph Laplacian:")
print(norm_laplacian_matrix)
```
